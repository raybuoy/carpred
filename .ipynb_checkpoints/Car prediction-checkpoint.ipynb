```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Car details v3.csv")
df.head()
```

```python
df.shape
```

```python
print(df["seller_type"].unique())
print(df["fuel"].unique())
print(df["transmission"].unique())
print(df["owner"].unique())
```

```python
df.isnull().sum()
```

```python
df = df.dropna(axis = 0)
df.shape
```

```python
df.describe()
```

```python
df.head()
```

```python
df.info()
```

```python
#preprocessing data with units and removing columns which are not required

df["engine"] = df["engine"].astype("str")
m =  df['engine'].str.contains('\d+ CC')
df['engine'] = df['engine'].str.extract('(\d*\.\d+|\d+)', expand=False).astype(float)
df['engine'] = df['engine'].div(1000)

df["max_power"] = df["max_power"].astype("str")
m =  df['max_power'].str.contains('\d+ bhp')
df['max_power'] = df['max_power'].str.extract('(\d*\.\d+|\d+)', expand=False).astype(float)

df["mileage"] = df["mileage"].astype("str")
m =  df['mileage'].str.contains('\d+ kmpl')
df['mileage'] = df['mileage'].str.extract('(\d*\.\d+|\d+)', expand=False).astype(float)

df.drop(["name"], axis = 1, inplace = True)
df.drop(["torque"], axis = 1, inplace = True)

df.head()
```

```python
df.describe()
```

```python
df["year"]=2021-df["year"]
df.head()
```

```python
#deal with categorical values fuel, seller_type, transmission, owner
df = pd.get_dummies(df, drop_first = True)
#df.head()
```

```python
df.corr()
```

```python
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize = (20,20))
g = sns.heatmap(df[top_corr_features].corr(),annot = True, cmap="RdYlGn")
```
